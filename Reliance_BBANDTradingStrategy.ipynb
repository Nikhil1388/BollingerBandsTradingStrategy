```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
```

```python
Reliance = yf.download("RELIANCE.NS", start = "2011-01-01", end = "2021-01-01")
Reliance
```

```python
plt.figure(figsize=(12,8))
plt.style.use("fivethirtyeight")
plt.plot(Reliance["Adj Close"] , label = "Reliance")
plt.xlabel("2011-01-01 to 2020-12-31")
plt.ylabel("Stock Price in INR")
plt.title("Reliance Stock Price History")
plt.legend(loc = "upper left")
plt.show()
```

```python
Reliance["SMA20"] = Reliance["Adj Close"].rolling(window=20).mean()
Reliance["Stddev"] = Reliance["Adj Close"].rolling(window=20).std()
Reliance["Upper BBAND"] = Reliance.SMA20 + 2*Reliance.Stddev
Reliance["Lower BBAND"] = Reliance.SMA20 - 2*Reliance.Stddev
```

```python
Reliance
```

```python
Reliance.dropna()
```

```python
Reliance["Buy_Signal"] = np.where(Reliance["Adj Close"] < Reliance["Lower BBAND"] , True, False)
Reliance["Sell_Signal"] = np.where(Reliance["Adj Close"] > Reliance["Upper BBAND"], True, False)
```

```python
plt.figure(figsize=(12,12))
plt.style.use("seaborn-deep")
plt.plot(Reliance[["Adj Close", "SMA20", "Lower BBAND", "Upper BBAND"]])
plt.scatter(Reliance.index[Reliance.Buy_Signal], Reliance["Adj Close"][Reliance.Buy_Signal], marker = "^", color = 'g', s = 250)
plt.scatter(Reliance.index[Reliance.Sell_Signal], Reliance["Adj Close"][Reliance.Sell_Signal], marker = "v", color = 'r', s = 250)
plt.fill_between(Reliance.index, Reliance["Lower BBAND"], Reliance["Upper BBAND"], color = "grey", alpha= 0.3)
plt.xlabel("2011-01-01 to 2020-12-31")
plt.ylabel("Price in INR")

plt.legend(["Adj Close", "SMA20", "Lower BBAND", "Upper BBAND"], loc = "upper left")
plt.show()                 
```

```python
buy_signal = []
sell_signal = []
open_pos = False

for i in range(len(Reliance)) :
    if Reliance["Adj Close"][i]<Reliance["Lower BBAND"][i] :
        if open_pos == False :
            buy_signal.append(i)
            open_pos=True
        
    elif Reliance["Adj Close"][i]>Reliance["Upper BBAND"][i] :
        if open_pos==True :
            sell_signal.append(i)
            open_pos = False
```

```python
plt.figure(figsize=(12,12))
plt.style.use("seaborn-deep")
plt.plot(Reliance[["Adj Close", "SMA20", "Lower BBAND", "Upper BBAND"]])
plt.scatter(Reliance.iloc[buy_signal].index, Reliance["Adj Close"].iloc[buy_signal], marker = "^", color = 'g', s = 250)
plt.scatter(Reliance.iloc[sell_signal].index, Reliance["Adj Close"].iloc[sell_signal], marker = "v", color = 'r', s = 250)
plt.fill_between(Reliance.index, Reliance["Lower BBAND"], Reliance["Upper BBAND"], color = "grey", alpha= 0.3)
plt.xlabel("2011-01-01 to 2020-12-31")
plt.ylabel("Price in INR")

plt.legend(["Adj Close", "SMA20", "Lower BBAND", "Upper BBAND"], loc = "upper left")
plt.show()      
```

```python
merged = pd.concat([Reliance["Adj Close"].iloc[buy_signal],Reliance["Adj Close"].iloc[sell_signal]], axis=1)
merged.columns = ["Buy","Sell"]
merged["Sell"] = merged["Sell"].shift(-1)
merged
```

```python
totalprofit = merged.Sell - merged.Buy
totalprofit
```

```python
totalprofit.mean()
```

```python
relprofit = ((merged.Sell - merged.Buy)/merged.Buy)*100
relprofit
```
